```python
# importing packages and reading csv files
import pandas as pd

import matplotlib.pyplot as plt

import sklearn

import numpy as np 

from sklearn.linear_model import LinearRegression



df = pd.read_csv('regrex1.csv')
```

```python
# displaying the data from the csv
df.head()
```

```python
# plotting the data
plt.figure(figsize=(8, 6))  # Optional: specify the figure size

plt.scatter(df['x'], df['y'], label = "Data Points", marker='o', color='blue', alpha=0.5)  # Plotting x vs y

plt.title('Data from regrex1.csv')  # Adding a title
plt.xlabel('x')  # Adding label to x-axis
plt.ylabel('y')  # Adding label to y-axis
plt.legend()

plt.grid(True)  # Optional: add gridlines
plt.tight_layout()  # Optional: adjust layout

plt.show()  # Display the plot
```

```python
# Setting up the linear model
model = LinearRegression()

# reshaping the data to be compatible with numpy
x = np.array(df['x']).reshape(-1, 1)
y = np.array(df['y'])
```

```python
# Model fitting
model.fit(x, y)
```

```python
# calcualting r square and printing intercept + slope
r_sq = model.score(x, y)
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
print(r_sq)

```

```python
# predicting the model
y_pred = model.predict(x)
```

```python
plt.figure(figsize=(8, 6))  # Optional: specify the figure size

plt.scatter(df['x'], df['y'], label = "Data Points", marker='o', color='blue', alpha=0.5)  # Plotting x vs y
plt.plot(df['x'],y_pred , label = "Fit")

plt.title('Data from regrex1 with slope plotted.csv')  # Adding a title
plt.legend()
plt.text (12,10,f'$R^2={r_sq}$')
plt.xlabel('x')  # Adding label to x-axis
plt.ylabel('y')  # Adding label to y-axis

plt.grid(True)  # Optional: add gridlines
plt.tight_layout()  # Optional: adjust layout



plt.show()  # Display the plot
```
